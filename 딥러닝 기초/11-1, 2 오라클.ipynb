```python
데이터 베이스를 관리해주는 DBMS를 오라클이라 한다. DBMS는 DB라고 한다. 
데이터 베이스 안에 여러개의 데이터 베이스가 있음 
파일은 버리고 데이터 베이스에 저장 
전역 베이에 DBMS 
관계형 데이터 베이스 - 테이블간 관계 (테이블 = 표 =릴레이션)
관계 -릴레이션쉽
sqlplus system/orcl
conn 커넥트 연결
system 이라는 강력한 사용자 밑에 scott라는 약한것 시스템이 맘대로 
CONN SCOTT/orcl 스콧 연결
ACCOUNT LOCK; 계정 잠구기

CONN SCOTT/orcl  계정잠궈서 안됨

ALTER USER SCOTT 계정변경
IDENTIFIED BY orcl 비번바꾸기
ACCOUNT UNLOCK; 계정 풀기 끝날때는 세미콜론 
DESC EMP; 
데이터 베이스 접근 스카마 접근
scott 사용자 접속 user가 사용자이면서 동시에 데이터 베이스이다. 안에 테이블있다. 


system에 들어가서 함 
ALTER USER HR
IDENTIFIED BY hr # 비밀번호 변경 
ACCOUNT UNLOCK;
empno not null은 무조건 널값이면 안됨
varchar2(10)  변할수 있는 캐릭터 2번 업그레이드 /가변형 문자/ 바차2/ 글자수 10개 
```
