```python
이름에서도 알 수 있듯, 평균을 이용해서 데이터들을 그룹화 시키는 건데요! 
어떤 원리로 하는 건지 한번 알아보시죠! 
K-means 알고리즘에서 제일 중요하다고 할 수 있는 개념이 바로 "중심점(Centroid)"입니다. 
먼저 그림으로 과정을 보여드린 뒤에 텍스트로 설명드리겠습니다. 그게 이해가 제일 빠르더군요. 
먼저 아무 곳에나 중심점 2개를 둡니다. 그럼 이 중심점으로 데이터들은 Grouping이 됩니다. 

위쪽에 있는 초록색 중심점과 가까운 데이터들은 초록 중심점 소속으로, 

아래에 있는 노란색 중심점과 가까운 데이터들은 노랑 중심점 소속으로 

각각 구분이 된다는 것이죠.


그러나 저 중심점의 위치가 과연 정말 최적의 위치일까요? 데이터들을 좀 더 잘 구분하려면 
"중심이 데이터들의 가운데"에 있어야 합니다. 

가운데로 가기 위해서, 중심점은 데이터와 본인과의 거리를 계산하면서 점점
데이터의 중심으로 이동하기 시작합니다. 
각 데이터와 중심점간의 거리를 나타낸 그림입니다. 
이 중심점까지의 거리가 평균적으로 비슷할 때까지 중심점을 이동시키는 것이죠.

중심점이 이동되면, 이동한 중심점에 대해 데이터들의 소속 또한 다시 새롭게 정의됩니다. 

지금의 그림에선 나타나지 않지만, 이동하기 전에는 초록색 중심점 소속이었지만,
이동하고 나서는 노란색 중심점 소속일 수도 있다는 것이죠. 
n_clusters 가장 중요한 파라미터입니다. 군집화를 몇개의 그룹으로 수행할 것인지를 명시합니다.
init 처음에 중심점을 둘 때 어떻게 둘 것인지를 명시합니다. 
보통 'k-means++'라는 알고리즘으로 설정합니다.
max_iter 중심점을 이동시키는 최대 반복 횟수입니다. 반복횟수 전에 이동을 완료하면 멈춥니다. 
위와 같이 군집화를 수행하고나서의 수행 결과는 labels_를 통해 확인하실 수 있습니다. 

-------------------------
평균 이동 군집화는 위에서 설명드린 K-Means와 상당히 비슷한 방법으로 진행됩니다. 
다만 이젠 좀 더 뭔가 수학적으로 군집화를 시키려는 시도들이 눈에 보이는데요! 

K-Means가 거리를 재서 중심으로 이동했다면, 
평균 이동은 "데이터들이 가장 많이 뭉쳐있는"곳을 중심점으로 잡습니다. 
즉, 데이터들의 밀도를 보고 중심점을 잡고 군집화를 수행한다는 얘기인데요, 
이 "밀도"를 어떻게 측정하는지 그 원리에 대해 살펴보도록 하겠습니다. 


먼저 평균 이동 군집화는 KDE(Kernal Density Estimation)를 이용합니다. 
이름에 Density가 있으니 밀도를 이용하는 것같긴 한데.. 이걸 어떻게 활용할 수 있을까요? 
통계의 목적은 표본(Sample)을 보고 모집단을 "잘" 추정하는 것을 목표로 합니다. 
때문에 샘플이 모집단의 성질을 얼마나 잘 반영하냐가 중요한 이슈가 되고, 
어떤 가설을 세웠을 때 그것이 말이 되는 가정인지 아닌지를 따지는 것도 그 때문이죠.

여기서 밀도추정 (Density Estimation)이란 sample 데이터들의 분포를 보고, 
모집단의 분포를 추정하고자 하는 시도입니다. 

예를 들어, '헬스장 이용객 수'라는 변수가 있을 때, 연초와 여름엔 이용객 수가 하루 1000명 이었지만 
가을 겨울엔 100이 되었다고 합시다. 그런 데이터가 10년치가 쌓이면 대강 헬스장 이용객 수가
어느 정도의 분포를 가지고 있는지 확인할 수 있습니다. 


이러한 밀도추정 방식은 크게 두가지 방법이 있습니다. 

바로 모수추정(Parametric)과 비모수추정(Non-Parametric)이 있는데요!

정말 간단하게 얘기해서

모수추정은 데이터의 분포가 특정 분포(모델)를 따르고 있다~고 되게 이상적으로 가정하면서
추정하는 방식이고

비모수추정은 정말 날것의 관측된 데이터만을 가지고 추정하는 방식입니다. 
여기서 비모수추정의 가장 대표적인 방식이 바로 히스토그램과 KDE입니다. 

히스토그램은 중학교때부터(요즘은 초등학교때부터 인가요,,?) 많이 봐온 정말 친숙한 그래프죠?
이런 오랜 친구를 놔두고 왜 KDE라는 이름부터 낯선 녀석을 쓰는 걸까요? 
그건 바로 히스토그램의 단점에서 기인합니다. 

같은 데이터 분포에 대해, 히스토그램은 bin의 크기에 따라 불연속적인 모양을 띌 수 있고, 
동시에 bin을 어떻게 조정하느냐에 따라 다른 밀도함수 모양을 가집니다.

그에 반면 KDE는 상당히 부드러운 곡선 모양의 확률밀도그래프가 그려짐을 확인할 수 있습니다. 
이러한 성질 때문에 히스토그램보다는 KDE를 사용하는 것이죠.


KDE는 각 데이터에 대해 커널 함수를 적용한 후에, 이들을 모두 합쳐서 최종 모양을 나타냅니다. 
여기서 커널함수란 "원점을 대칭으로 적분값이 1이 나오게 하는 함수"입니다. 

이렇게 매~~우 많은 함수들이 있지만!! 우리는 여기서 가장 대표적인 가우시안 커널함수를 사용합니다. 
(보면 가우시안을 되게 많이 쓰는데 왜 그런지는 따로 알아봐야겠네용 ㅎㅎ )


아무튼 각 데이터에 모두 이런 식으로 커널함수를 적용한 다음, 모두 더해주고 데이터 건수n으로 나눠주기만 
하면 밀도함수를 구할 수 있다~ 라는 것이 바로 KDE입니다. 

식으로 나타내면 아래와 같습니다. 

여기서 하나 중요한 개념이 나오게 됩니다. 바로 대역폭(Bandwidth)라는 개념입니다. 

수식에선 h라고 표시되어있습니다. 



대역폭은 각각에 적용할 커널함수를 얼마나 부드럽게 나타낼것인지?를 나타냅니다. 
경영수학 시간에 배우는 지수평활법에.. 알파와 같은 존재라고 해도 될것 같습니다. 


위에서 설명 드렸다싶이, 각 데이터에 우리는 커널함수를 적용합니다. 
커널함수는 적분값이 1이 나오게 하는 원점이 대칭으로 만드는 함수인데, 
이 1을 만들 때 "뾰족하게" 만들것이냐? "부드럽게" 만들것이냐?를 결정합니다. 

예상하신 대로, Bandwidth가 작으면 뾰족하게 나타나며, 클 수록 부드럽게 나타납니다. 



Bandwidth가 매우 작게 되면.. 그만큼 데이터를 나타내기가 쉬워지지만 우리가 항상 경계하는 
Overfitting이 나타나기 매우 쉽습니다. 그러나 너무 크게 설정하는 경우엔 
오히려 Under Fitting이 되기 쉽죠.


그럼 자연스럽게 생각의 흐름을 따라가면, 각각의 커널 함수 모양을 결정하는 Bandwidth를 어떻게 조정하느냐에
따라서 KDE의 모양이 달라질 것을 예상하실 수 있습니다. 


Bandwidth를 작게 설정하면 데이터 각각의 커널이 상당히 뾰족하게 나타날 것이고, 
이를 합산한 KDE 또한 매우 많은 봉우리를 가질 것입니다.
즉, 밀도가 높은 부분이 많이 나타나게 될 것입니다.

반면 Bandwidth를 크게 설정하면 각각의 커널이 전체적으로 평활하게 나타날 것이며,
KDE 또한 상대적으로 적은 봉우리를 가질 것입니다. 



Mean Shift는 이 봉우리(군집중심점)를 가지고 군집화를 수행합니다. 때문에 

"군집화를 하기에 가장 적절한 Bandwidth값"을 아는 것이 상당히 중요합니다. 



물론 우리의 위대한 통계학자들은 최적의 Bandwidth 값을 구하는 공식을 이미 마련해놨습니다. 

수많은 KDE 함수 중, Gaussian을 사용하는 경우의 최적 Bandwidth는 다음과 같습니다.

먼저 sklearn.cluster 에서 MeanShift를 import 해줍니다. 
그러고 나서.. Bandwidth값을 설정한 객체를 생성해주고, 
여느 사이킷런과 마찬가지로 fit_predict를 해주시면 되겠습니다.


먼저 Bandwidth를 0.8로 설정한 경우의 군집화 수는 012345678

이렇게 9가지가 나오는 것을 확인하실 수 있습니다. 
애초에 생성한 데이터가 군집화가 3개가 되도록 만든 것이기 때문에 Bandwidth를 늘려 조정해보겠습니다.
이번엔 Bandwidth=1로 설정한 결과입니다. 

저희가 원하는 대로 딱 3개의 군집이 형성되었음을 확인하실 수 있습니다! 



근데 맨날 이렇게 수작업으로 확인해야할까요? 아까 분명 위에서 그렇게 위대하신 통계학자님들께서 
공식을 마련해놨네 뭐네 하더니.. 하핫.. 그럴줄 알고 최적의 bandwidth값을 제공하는 함수가 따로 있습니다. 
사이킷런의 estimate_bandwidth() 를 사용하시면 군집화에 맞는 최적의 bandwidth값을 제공해줍니다 !



그럼 이렇게 최적의 bandwidth값도 구했으니.. 군집화를 수행한 후의 결과를 시각화 시켜보면 다음과 같습니다. 
이렇게 3개의 군집으로 잘 군집화가 수행됨을 확인하실 수 있습니다. 

Mean Shift의 장점이라고 하면, 거리에 기반하는 것이 아닌, 정말데이터들이 모인 밀도를 가지고 
판단하는 것이기 때문에 K-means보다 더 유연한 군집화가 가능합니다. 그러나.. 데이터 하나하나에 
커널을 적용하고 합치고~ 뭐 이런 과정이 솔직히 짧게 걸리진 않습니다. 때문에 시간 소요가 조모 
걸리고 bandwidth가 조금만 변해도 군집화가 확 변한다는 단점이 있습니다.(너무 예민하죠) 


때문에 일반적으로 사용되진 않고,, 컴퓨터 비전과 같은 이미지나 영상 데이터에서 객체를 구분하는 분야에
많이 활용된다고 합니다 ! 


----------------------------------------------------------------


GMM은 기본적으로 우리가 가진 데이터들이 모두 정규분포(Normal Distriburion)을 따른다고 가정하고 
시작합니다. 때문에 위에서 배운 평균이동이 이러한 가정을 하지 않은 비모수 추정이었다면,
GMM은 모수추정이라고 할 수 있습니다! 
GMM의 원리를 알기 위해선 또 우리가 알고 가야할 통계학의 방법론이 있습니다.

바로 '최대우도법'과 'EM알고리즘' 입니다! 차근차근 하나하나 알아가보도록 하겠습니다.


최대우도법(Maximum Likelihood Estimation)

최대우도법은 데이터의 분포를 보고 이 데이터의 모수는 어떤 분포를 따를 것인가? 를 예측하는 것입니다.

데이터가 이렇게 분포가 되어있다고 할 때, 왼쪽의 데이터는 파란색의 분포를 따르는 것이 맞아보이고,
오른쪽의 노란 데이터들은 노란색 분포를 따르는 것이 맞아보이죠. 
이런식으로 데이터를 보고 얘네들의 모수는 어떤 분포를 따를까? 를 맞추는 것이 바로 최대우도법입니다. 


그럼 여기서 이제 알아야하는 것은 어떤 모양의 분포를 따를 것인가? 가 우리의 관심사가 됩니다.

그런 모양을 가지는 매우 다양한 분포들이 있을 테니까요. 가장 적절한 분포는 뭘까? 하는 겁니다. 

이때 쓰이는 것이 바로 우도(Likihood)입니다. 

정~~말 간단히 얘기하면, 각 데이터가 해당 분포에 있을 가능성(가능도)를 의미하죠. 

그럼 최대 우도법이라고 하면... 이 가능도를 최대화 시키는 것을 의미한다는 건데.. 
가능도를 저 그림에서 어떻게 나타내죠? 

바로 분포까지의 높이로 보고 계산합니다. 

각 데이터들과 분포까지의 높이를 모두 곱한 값이 최대가 되는 분포를 구합니다. 

다르게 생각하면, 데이터를 가장 잘 설명할 수 있는 가능성이 제일 높은 확률분포를 구하는 과정이죠.

이를 수식으로 표현하면.. 다음과 같습니다.  

방법은 좋습니다. 각 데이터들이 어떤 정규분포에 속해있는지 가장 확률이 높은 정규분포로 소속을 정한다는
것인데... 이를 현실적으로 적용할 때 한 가지 문제가 발생합니다.

바로 우리는 데이터들'만' 가지고 있는 겁니다. 위의 예시에선 라벨(데이터들의 정답값)이 있어서 분포만 
어떻게 생겼는지(평균과 분산이 어떻게 되는지!)만 알면 됐지만, 이젠 그 라벨조차 없는 것이죠.

정리하면, 라벨이든 분포든 뭐든 하나라도 알아야 제대로 소속을 정할 수 있는데... 둘 다 없어요!! 
우리는 또 이래서 노가다를 해야합니다. 어차피 컴퓨터는 미지의 영역을 찾기 위해 딱 계산된 식을 노가다 하는
것때문에 사용하죠... 

아무튼 라벨과 분포 중 어느 값을 랜덤하게 아무거나 설정한 뒤에, 그 랜덤한 것에 대해 다른 하나를 추정해보고
, 또 수정하고.. 이런 과정을 반복합니다.

GMM에서는 라벨을 미지의 것으로 두고, 분포를 랜덤하게 산정한 후, 맞춰나가는 방식을 사용합니다. 

방식은... 그림으로 설명하겠습니다.  

먼저 우리의 모델이 마주한 데이터는 이런 상황입니다. 어디에 속해있는지 라벨도 없고, 
소속값을 계산할만한 분포도 없습니다(평균, 분산도 없습니다). 그래서 일단... 
라벨과 분포 중 분포를 아무 값이나 랜덤하게 설정해봅니다. 


이렇게 그냥 진자 아무거나 분포를 설정합니다. 평균과 분포가 제멋대로죠.

그럼 이제 라벨을 정할 수 있게됩니다. 분포가 정해지면 라벨을 정하는 건 높이만 비교하면 된다고 
말씀 드렸습니다. 예를 들어,
가운데의 겹치는 부분의 데이터는 초록색과 주황색 중, 어느 분포에 속하게 될까요? 
네. Likelihood를 따라가면 높이가 좀 더 높은 초록색 분포에 속하겠구나~로
"일단 지금은" 예상 가능하다는 것이죠. 자연스럽게 두 데이터는 초록색 분포에 속하게 됩니다.

그럼 여기서 끝이 나고 모두가 하하호호 행복한 상황에 있게 될까요? 
아니요 ㅠㅠ 아쉽게도 우리는 좀 더 해야합니다. 



최대우도법을 설명해드릴 때, 처음에 라벨을 보고, 모두 곱한 값이 최대가 되게하는 분포를 정할 수 있다고
했습니다. 이제 라벨을 어느 정도 정했으니, 분포를 새롭게 다시 그려볼 차례가 된 것이죠. 

이렇게 정리된 라벨을 따라 분포를 어느 정도 다시 그려줍니다. 그럼 또 다시 우리는 라벨을 높이
비교를 통해 새롭게 결정할 수 있겠죠?

이런 식의 과정을 '더 이상 변하지 않을 때까지' 반복해줍니다.



여기까지가 GMM의 대략적인 과정입니다. 



EM알고리즘 또한 결국 우리가 위에서 한 과정을 반복하는 것에 불과합니다. 
마음같아선 여기에 찬란한 수식들을 적고 싶지만 어차피 저도 이해하지 못할 거 그러는 게 의미가 있나 싶네요.. 아직 수식을 통한 이해를 할 수 있는 레벨은 도달하지 못한 것같습니다.. 문돌이에겐 너무 가혹하네요 ㅠㅠ 

E(Expectation) 과정

개별 데이터가 어떤 정규분포에 속할지 결정하는 과정

M(Maximization) 과정

각 그룹의 모수(평균,분산)을 재계산 하는 과정


"응? 이거 우리 아까 했던 거 아니야?" 

맞습니다!!!! EM알고리즘 과정을 계속 반복하면서, 
그 안에서 최대우도법을 이용하여 군집화를 수행하는 것이 GMM이라고 정리할 수 있겠습니다!!



그럼 이제 작동원리를 알았으니, 실전에서 어떻게 사용하는지를 알아봐야겠죠? 

이러한 GMM 군집화는 우리가 앞서 배운 K-Means 군집화의 한계를 보완해줍니다.



K-Means는 길쭉하게 생긴 분포를 가진 데이터들을 군집화 시킬 때 상당히 불리합니다. 
애초에 군집을 생성하는 방법이 원형을 이루면서 가기 때문이죠.

반면, GMM은 길쭉한 모양의 분포를 가진 데이터 또한 상당히 잘 군집화시키는 것을 확인할 수 있습니다.
직접 그 차이를 확인해보겠습니다. 

----------------------------------------------------------
먼저 이렇게 분포되어있는 데이터가 있다고 해봅시다. DBSCAN은 먼저 우리가 익숙한 개념인
"핵심 포인트(Core Point)"를 지정합니다. 

핵심 포인트를 지정하는 기준은 주변 영역에 일정 수 이상의 데이터가 있으면 핵심포인트입니다.

한마디로... 데이터 중에서 인싸를 고른다고 하면 이해가 편합니다! 

예를 들어,,, 빨간색으로 표시한 데이터의 주변 반경으로는 자신을 포함해서 6개의 데이터를 가지고 있습니다. 

인싸의 기준을 "반경 100m에 본인 포함 친구 6명 이상 있는 사람"으로 정의하면,
저 빨간색 데이터는 "인싸"로 지정된 것이죠. 



이 인싸가 바로 핵심포인트(Core Point)이며, 반경 100m처럼 데이터의 얼마만큼을 주변으로 잡을 것이냐를
정한 것을 엡실론(Epsilon)으로 부릅니다. 

핵심포인트 하나를 찾으면, 이제 다른 핵심 포인트를 찾을 차례입니다.
초록색으로 표시한 데이터 또한 주변에 본인을 보함한 6개의 데이터가 있으니, 인싸, 즉, 핵심포인트 입니다. 

이렇게 같은 엡실론 영억을 공유하는 두 군집은 같은 군집으로 분류됩니다. 

빨간색으로 표시된 핵심포인트를 중심으로 한 영역과, 초록색으로 표시된 핵심포인트를 중심으로 
한 영역은 이제 같은 소속, 군집이 된 것이죠. 

그럼 노란색으로 표시한 데이터 같은 경우는 어떻게 할까요? 노란색 데이터 같은 경우엔,
주변에 6개 이상의 데이터가 없습니다. 아쉽게도 5개라 인싸는 못돼죠... 
그럼 노란색 데이터는 핵심포인트가 되지 못합니다.

노란색데이터과 같은 데이터는 "경게 포인트(Border Point)" 라고 합니다.
군집화의 경계선을 긋는 역할을 하죠.
그럼 검정색으로 표시된 데이터는 뭘까요? 네... 저들이 바로 "아싸"입니다 ㅠㅠ 

저런 검정색 데이터처럼 핵심포인트도 근처에 없고... 군집화 반경 바깥에 있는 데이터를
"잡음 포인트(Noise Point)" 라고 합니다. 아싸죠 아싸...... 킄......................

이런 식의 잡음 포인트들은 자동으로 제거됩니다.  알아서 Outlier를 탐지해서 지워버리는 것이죠!! 

DBSCAN은 이와 같은 방법으로

군집갯수를 알아서 생성하고, outlier를 자동으로 지워줍니다.
다만, 밀도가 너무 다 비슷하거나 피쳐가 많은 경우엔 성능이 저하되는 모습을 보입니다.


그럼 이렇게 직관적인 알고리즘을 가진 DBSCAN은 어떻게 사용할까요?
그 사용법을 지금부터 한번 알아보겠습니다 ! 
min_samples - 핵심포인트가 되기 위해 엡실론 반경 내에 포함되어야 할 최소한의 데이터 갯수입니다.
본인을 포함한 갯수입니다.

metric - 반경을 계산할 때, 어떤 방식으로 계산할 것인지를 지정합니다. 보통 유클리드 거리를 사용합니다.
DBSCAN은 기하학적인 분포를 가지는 데이터 셋에 매우 유리합니다. 예를 들어 동그랗게 생긴 데이터가 있죠.

도넛처럼 생긴 데이터를 우리가 위에서 배운 군집화 알고리즘을 적용해보고, 
어떤 알고리즘이 제일 적합한지 확인해보도록 하겠습니다. 

먼저 K-Means의 경우입니다.



제대로 분류되지 않은 모습입니다.. 거리를 기반해서 그런지, 위 아래로 군집화를 수행해버렸습니다.



GMM 또한 길쭉한 데이터셋에 특화된 알고리즘이라 그런지... 대각선으로 군집화를 수행한 모습입니다. 
여전히 잘 분류해내지 못했습니다 ㅠㅠ 

------------------------------------------------------------------

군집 평가(Cluster Evaluation)

우리가 예전에 배운 분류에서는, 모델의 성능을 평가할 때 정답값과 모델의 예측치를 비교해가면서
"채점"을 했습니다. 그러나 군집화는 비지도학습답게 정답값이 제공되지 않죠. 

때문에 군집화를 평가할 때는 새로운 평가 지표가 필요합니다. 

바로 실루엣 분석(Shilhouette analysis) 입니다! 



실루엣 분석은 생각보다 그 원리가 간단합니다. 우리가 지난번에 한 차원축소에서 , 
LDA의 원리를 기억하시나요? 피쳐들을 압축하는 과정에서 분류까지 수행하는 알고리즘이었죠! 
아래 그림을 보시면 기억이 날 수 있습니다.
이런 식으로 같은 군집 내의 데이터끼리는 얼마나 잘 가깝게 모여있고,
다른 군집의 데이터와는 얼마나 멀리 떨어져 있는지가 중요한 성능 지표가 됩니다. 



위의 말을 좀 더 수식으로 표현하기 쉽게 말을 정리하면 다음과 같습니다.

군집(a)에서 a(i)를 "i번째 데이터에서 자신이 속한 군집 내의 다른 데이터포인트들의 평균 거리"로
정의합니다.

즉, a(i) = mean(a12, a13, a14....)


반면, b(i)는 "i번째 데이터에서 가장 가까운 타 군집내의 다른 데이터포인트들의 평균 거리"로 정의합니다.

즉, b(i) = mean(b15, b16, b17, b18....)

두 군집간의 거리는 b(i) - a(i)로 표현될 수 있는 것이죠.
max(a(i), b(i))로 나눠주는 이유는 정규화를 하기 위함입니다. 
(사실 군집간의 거리를 나타내는 b(i)로 나눠야 맞는데.. 실루엣 계수의 값을 -1,
 즉, 잘못된 경우도 포함시켜주기 위해 a(i)도 조건에 포함시켜줍니다.)



이렇게 나온 실루엣 계수(Silhouette Coefficient)는 -1~1 사이의 값을 가집니다.

-1일 수록 좋지 않은 것이며, 1일 수록 군집화가 잘 수행된 것입니다.

0이 서로 다른 군집끼리 가까운 것을 의미하는데, 실루엣 계수가 -값을 가진다는 것은.. 
군집끼리 서로 영켜있다~~ 라는 말이 되겠죠. 상당히 좋지 않은 경우입니다. 


이런 식이면, 동떨어진  파란색 군집의 실루엣 계수는 매우 높게, 잘 나올 것입니다.

그러나  겹쳐진 주황색 부분의 군집들은 실루엣 계수가 나쁘게 나오게 나오겠죠.

이들의 평균을 구해보면 그럭저럭 꽤 괜찮은 실루엣 계수가 나와버릴 것입니다.


이런 경우가 있을 수 있기 때문에, 이 점을 유념한 채로 계수를 잘 봐야한다~~ 는 것입니다! 


그럼 실루엣 계수는 어떻게 분석할까요? 


silhouette_samples는 각 데이터의 실루엣 계수를 반환합니다.

silhouette_score는 silhouette_samples를 통해 나온 전체 데이터의 실루엣 계수의 평균값을 구해줍니다.

이런 식의 그림이 나옵니다. 위의 그림 같은 경우엔 평균 실루엣 계수가 0.7정도로 매우 높게 나왔습니다만,
초록색 군집이 너무 뛰어나서, 회색 군집의 좋지 않은 군집화를 커버해서 나타나는 평균치였습니다.


반면, 4개의 군집으로 만들었더니, 전체 평균 실루엣 계수도 나쁘지 않게 나오고, 
각 개별 데이터의 실루엣 계수 또한 평균 실루엣 계수에서 크게 벗어나지 않았습니다.
고루고루 모두 잘 하는 매우 훌륭한 팀인 것이죠.

이런 경우, 비로소 매우 적절한 군집화가 이루어졌다고 할 수 있겠습니다.



```
