````python
2. Vectorization

저희가 지난 NLP시간에 배웠던 Vectorize 기억나시나요? 
One-Hot Encoding, Embedding 등, 텍스트를 의미를 가지는 벡터로 만들기
위한 다양한 방법들이 있었습니다. 이번 데이터에서는 장르를 기반으로 할 것이기 때문에,
복습겸 NLP 때 배웠던 CountVectorizer로 Text Vectorization을 진행하겠습니다. 
Tfidf로 하지 않는 이유는 텍스트의 종류가 딱히 많지 않기 때문이죠 (만약 리뷰를 기반으로 했었으면
Tfidf로 했을 겁니다)
````

````python
from sklearn.feature_extraction.text import CountVectorizer

## 영화의 장르마다 띄어쓰기로 구분한 새로운 column 생성
movies_df['genres_literal'] = movies_df['genres'].apply(lambda x : (' ').join(x))

## Vectorization 실행
count_vect = CountVectorizer(min_df = 0 , ngram_range=(1,2))
genre_mat = count_vect.fit_transform(movies_df['genres_literal'])

````

````python
```
이렇게 하시면 기존의 DataFrame에 genres_literal 이라는 이름의 새로운 column이 추가되었습니다. 그 안의 내용은 우리가 의도하던 대로 genres list 내부의 요소들이 띄어쓰기로 구분되어있죠. 

이제 위에서 CountVectorizer로 구한 genre_mat 으로 Cosine Similarity를 구해보겠습니다.

(cosine Similarity는 기억나시죠? 두 벡터가 얼마나 비슷한 방향을 향하고 있는지를 통해 유사성을 측정하는 지표였죠!!)
```
````

````python
# 3. Cosine Similarity 측정
from sklearn.metrics.pairwise import cosine_similarity

genre_sim = cosine_similarity(genre_mat, genre_mat)
print(genre_sim)

#1번째 원소는 본인-본인의 비교니까 당연히 1이고, 2번째 요소는 첫번째 행과 두번째 행과의
#cosine similarity,, 그 옆은 첫번째 행과 세번째 행과의 cosine smiliarity가 되겠죠? 

#즉, genre_sim 이라는 Matrix는 영화들끼리 비교를 해봤을 때,
#장르가 얼마나 유사한지를 나타낸 Matrix가 됩니다.
````

````python
4. 기준에 의한 추천


이렇게 나온 애들을 sorting 해주기 위해서는 그냥 sort기능을 이용해도 되지만, 
저희는 이걸 가지고 좀 더 다양한 것들을 진행하기 위해 sort가 되게하는 index를 구할겁니다.
그런 기능을 하게 해주는 것이 바로 numpy의 argsort()입니다.
a 라는 객체에 저장된 array 가 오름차순으로 정렬되려면 [0,1,2,3,4,5]가 돼야할 겁니다. 
argsort()는 [0,1,2,3,4,5]가 되게하는 원래 array (a) 의 index를 반환해줍니다.



즉, [0,1,2,3,4,5]가 되려면... [0]번째에 있는 애가 처음, [3]번째에 있는 애가 그 다음,
[1]번째에 있는 애가 그 다음.... 이런 식으로 [0,3,1,4,2]라는 index array를 반환해준다는 것이죠. 


저희는 cosine similarity가 높은 순서대로 영화를 추천해줄 예정이니, 오름차순 index가 아니라 
내림차순 index를 받는 것이 좋겠죠. argmax()[::-1]를 해주시면 그냥 거꾸로 뒤집어줍니다. 
오름차순이었던 것이 내림차순이 되죠. 다만 우리는 2차원 행렬를 다뤄야하므로, argsort[:,::-1]을 
해줍니다. 모든 행에 대해서 내림차순 정렬을 수행해주는 것이죠.
이렇게 정렬된 index list도 구해냈으니, 이제 이걸 바탕으로 어떻게 추천을 할까~ 를 고민하면 됩니다. 



근데 단순하게 cosine similarity 로만 추천을 해주면 문제점이 하나 발생합니다.
오직 장르가 얼마나 비슷하냐에 따라서만 추천을 해주기 때문에, 추천된 영화가 좋은 영화인지 아닌지
전혀 고려하지 않는다는 것입니다.

때문에 우리는 추가적인 요소를 한 가지 더 고려하고자 합니다. 바로 가중평점(Weighted Rating)이라는 
요소입니다. 가중평점 식은 다음과 같습니다. 
여기서 

V : Vote Count (투표 횟수)

M : 평점 부여를 위한 최소 투표 횟수

R : Vote Average (평균 평점) [개별 영화]

C : 전체 영화에 대한 평균 평점  [전체 영화]

저희는 함수를 하나 만들어서, 함수에 제목을 입력해주면 그 영화와 장르
Cosine Similarity가 높으면서 Weighted Rating도 함께 높은 순서대로 정렬하는 걸 구현해보겠습니다. 
````

````python
from sklearn.metrics.pairwise import cosine_similarity

genre_sim = cosine_similarity(genre_mat, genre_mat)
print(genre_sim)
genre_sim_sorted_ind = genre_sim.argsort()[:,::-1] 
percentile = 0.6
m = movies['vote_count'].quantile(percentile) ## 전체에서 60%에 해당되는 수를 최소 투표 횟수로 지정
C = movies['vote_average'].mean() ## 전체 영화에 대한 평점

def weighted_vote_average(record) : 
    v = record['vote_count']    ## 들어온 DataFrame에 대해 투표 횟수로 객체 저장
    R = record['vote_average']  ## 들어온 DataFrame에 대해 개별 영화 평균
    
    return ( (v/(v+m))*R + ((m/m+v))*C )  ## 위의 수식을 Return

## 기존 DataFrame에 weigted_vote_average 칼럼을 추가해줌
movies_df['weighted_vote'] = movies.apply(weighted_vote_average, axis=1)
def find_sim_movie(df, sorted_ind, title_name, top_n = 10) :
    title_movie = df[df['title'] == title_name]
    title_index = title_movie.index.values ## 해당 영화의 index 반환
    
    ## 여기가 바로 argsort를 쓴 이유입니다!! 
    similar_indexes = sorted_ind[title_index, 1 : (top_n*2)] ## 해당 영화에서 가장 유사한 top_n*2개의 영화 추출 
                                                            ## 대신 자기 자신은 제외 !! 
    similar_indexes = similar_indexes.reshape(-1) ## 2차원으로 되어있으므로 1차원으로 바꿔줌 (한줄짜리로!!)
    
    ########## 이러면 similar_indexes 에는 후보군의 index가 담겨있음 ##########
    
    final_list = df.iloc[similar_indexes].sort_values('weighted_vote', ascending=False)[:top_n]
                ## Weighted Vote를 기준으로 내림차순 정렬하고 top_n 만큼만 가져옴 !! 
    return final_list
    
find_sim_movie(movies_df, genre_sim_sorted_ind, 'Avatar', 5)
````

````python
최근접 이웃 협업 필터링 추천 시스템

(Collaborative Filtering Recommend System)

최근접 이웃 협업 필터링의 예시는 여러분이 어떻게 보면 제일 익숙하실 겁니다. 
쿠팡같은 온라인 쇼핑물에 들어가서 상품을 볼 때, 다음과 같은 문구를 한 번 쯤은 보신 적이 있죠?


"이 상품을 고른 사람은 다음과 같은 상품도 구매했습니다." 

"당신과 비슷한 사람들은 다음과 같은 상품을 구매했습니다."


이렇듯, 나와 비슷한 사람들이, 혹은 내가 고른 아이템과 비슷한 다른 상품이 뭔지를 보면서 추천을 하는 시슽템이 바로 최근접 이웃 협업 필터링입니다. 이해하기 쉽게 비유를 하자면 

새로 개봉한 영화를 보기 전에, 나와 취향이 비슷한 친구에게 영화가 어땠냐고 물어보는 것과 같죠. 

취향이 비슷한 만큼, 그 영화에 대한 평가가 비슷할 테니까요.


위에 예시로 들어드린 문장은 각각 다음의 기술을 의미합니다.


1. 이 상품을 고른 사람은 다음과 같은 상품도 구매했습니다.   →  아이템 기반(Item-Item)

2. 당신과 비슷한 사람들은 다음과 같은 상품도 구매했습니다. → 사용자 기반(User-User) 


각각이 어떤 방식을 이용하는지 살펴본 다음에 실습도 진행해보겠습니다.

사용자 기반 최근접 이웃 방식

사용자 기반은 말 그대로 나와 비슷한 사람은 이 상품에 대해 어떻게 평가했는가? 를 참고한 겁니다.


내가 고르지 않은 상품에 대해 나와 물품 구매 취향이 비슷한 B라는 사람이 이미 골랐다면
B가 어떻게 그 상품을 평가했는지를 보고 마음에 들어할지 안 들어할지 예측한다는 것이죠.


쉽게 그림으로 설명드리면
A 부터 D라는 사람이 있고, 3대 측정 종목인 벤치, 데드, 스쿼트가 있다고 해봅시다.

이 3개의 운동에 대해 4명이 메긴 점수를 나타낸 Matrix입니다. (애초에 이렇게 표로
                                        나타내는 작업부터 해야합니다 ㅋㅋ)

​
자, 그럼 이제 ? 에 뭐가 와야할 지 예측해야하는 과제가 주어졌는데, 여러분들
같으면 어떻게 할 것같은가요?

대충 표를 봤을 때, 

A와 B는 상체 운동 위주를 좋아하는 것 같네요( 데드는 전신이지만 뭐 루마니안이라고 해봅시다 ) 
. A와 비슷한 운동 취향을 가진 B가 하체 운동인 스쿼트의 점수를 매우 낮게 평가했습니다. 
하체가 힘들다고는 하지만 좀 너무하네요 ㅠ ㅋㅋㅋㅋ 아무튼 A와 비슷한 취향의 B가 스쿼트의
점수를 낮게 줬으니, A도 스쿼트에 대해 낮은 점수를 줄 것이라고 예측합니다.



이번엔 D를 채워볼까요? C와 D는 A,B와는 다르게 스쿼트를 진짜 좋아하는 모습을 보입니다. 
이와 비슷한 취향을 가지는 D도 C처럼 벤치프레스의 점수를 낮게 주었을 것이라고 예측하는 것이죠.

간단하죠? 이제 사용자 기반 최근점 이웃 방식에 대해 알아봤고,
아이템 기반은 어떻게 하는지 알아보겠습니다.


아이템 기반 최근접 이웃 방식

사실 아이템 기반도 사용자 기반과 비교해서 방식이 크게 다르진 않습니다. 
사용자는 비슷한 사용자가 뭘 좋아했냐고 보는 거고, 아이템 기반은 사용자들이 평가를 
비슷하게 내린 아이템을 추천하는 것이죠.


말만하면 어렵습니다. 한번 그림으로 쉽게 나타내보겠습니다.



아까와 다르게 이제 왼쪽에 종목이 들어왔고 column에 사용자가 들어왔죠? 
자, 이번엔 새로운 종목인 레그프레스가 들어왔습니다. 



위에서 아이템 기반은 아이템에 대한 사용자들의 평가의 분포를 보고 평점을 예측한다고 했습니다.
"어떤 상품을 좋아할까?"에 좀 더 초점을 맞췄다고 생각하시면 편합니다.

벤치프레스는 10,9,8 // 데드리프트는 10,8,10 // 스쿼트는 2,3,3 입니다.
그럼 레그프레스는 사용자 A가 좋아할까요? 아님 싫어할까요? ​

A는 아마 싫어할 것입니다. 아이템에 대한 평가가 비슷한 스쿼트와 비교해봤을 때,
스쿼트에 대해 낮은 점수를 부여했기 때문이죠. 따라서 적어도 A에게 레그프레스는 추천되지 않을 것입니다.


그럼 C는 데드리프트를 좋아할까요? C에게 종목을 추천해줄 수 있다면, 어떤 종목을 추천할 건가요? 


아이템들에 대한 평점 분포를 살펴봤을 때, C는 벤치프레스와 가장 평점 분포가 유사한 데드리프트를
좋아할 확률이 높습니다. 따라서 C에겐 스쿼트나 레그프레스를 추천하는 것이 아닌, 
데드리프트를 추천하게 되는 것이죠.



아이템 기반은 아이템이 가지는 특성이나 속성이 비슷한 걸 추천하는 것이 아닙니다.
그건 오히려 컨텐츠 기반에서 우리가 이미 다뤄봤죠. 아이템 기반은 위처럼 "아이템에 대한 평가 분포"가
비슷한 아이템을 추천하는 방식입니다. 헷갈리지 않게 조심하세요!! 


현업에선 사용자 기반보다 아이템 기반을 더 많이 사용한다고 합니다.
위의 행렬들은 대부분 Sparse Matrix로 나타나게 되는데, 사용자 기반으로 비교를 
하기엔 정보가 많이 부족해서 그런 것이라고 하네요. 비슷한 제품을 좋아한다고 해서 꼭
그 사람들의 취향이 꼭 비슷하다고 단정 지을 수도 없습니다. (아이폰만 구매한 사람들끼리 영화 
                                   취향까지 비슷할 수는 없겠죠)


실습을 진행하기 위해 새로운 데이터셋을 가져오겠습니다. 

사용자 기반이나 아이템 기반을 하기 위해서는 위에서 말씀 드린 대로

"유저 - 아이템 간의 행렬"이 필요합니다. 즉, 어떤 유저가 어떤 아이템에 대해 어떻게 
평가했는지에 대한 데이터가 필요합니다. 우리가 위에서 이미 진행했던 데이터셋은 장르에 
대한 전체적인 평가만 있기 때문에, 협업필터링에 적용하기엔 무리가 있습니다. 때문에 아래의
데이터셋으로 진행합니다. 
````

````python
## movie id 를 표시하기 위해 다른 DF 를 가져옴
rating_movies = pd.merge(ratings, movies, on='movieId')

## 새로운 df로 pivot_table
ratings_matrix = rating_movies.pivot_table('rating', index='userId', columns='title')

## transpose()를 이용하여 index에 아이템, column에 user가 오게 만듦
ratings_matrix_T = ratings_matrix.transpose()

## NaN값이 있으면 곤란하므로 평가를 아직 하지 않았다는 의미에서 NaN을 0으로 바꿈
ratings_matrix_T.fillna(0, inplace=True)
ratings_matrix_T.head()
````

````python
간단하죠? 이제 우리가 원하는 방식대로 데이터를 준비 완료했습니다.


자, 아까 위에서 아이템 기반 방식은 평점 분포가 비슷한 아이템을 기준으로 추천을 한다고 했습니다.
그럼 이 평점 분포가 비슷하다는 것을 어떻게 컴퓨터에게 인식시킬 수 있을까요? 네.. 맞습니다. 
여기서 바로 Cosine Similarity를 사용합니다!


Cosine Similarity는 Text에만 사용할 수 있는 것이 아닙니다. 벡터로만 나타낼 수 있다면 
뭐든지 그 유사도를 측정할 수 있죠. 


위의 DataFrame을 가지고 Cosine Similarity를 측정해보겠습니다. (많이해서 익숙하시죠?)
````

````python
from sklearn.metrics.pairwise import cosine_similarity
## cosine 유사도 구하기
item_sim = cosine_similarity(ratings_matrix_T, ratings_matrix_T)

## 이걸 DF로 만들기
item_sim_df = pd.DataFrame(item_sim , index=ratings_matrix.columns, columns = ratings_matrix.columns)
item_sim_df.head()
````

````python
사용자 u의 아이템 i에 대한 예측 평점 값

Si,N
아이템 i와 가장 유사도가 높은 N개의 아이템 유사도 Vector (Cosine Similarity!!)

R_{u,N}
사용자 u의 아이템 i와 가장 유사도가 높은 N개의 아이템에 대한 실제 평점 (Real Rating)

{S}_{i,N}
그냥 정규화를 위해 나눠줌 (0~1 사이의 값으로 만들기 위해)


위의 요소들을 가지고 식을 다시 보면, 결국 실제로 주어진 Rating에 아이템 유사도라는 
가중치를 부여한 것입니다. (행렬끼리의 곱이니 내적을 해야겠죠?) 그렇게 하면,
실제로 부여한 Rating에 대해 유사도에 의한 영향력을 부여할 수 있습니다. 

즉, 이렇게 나온 결과들을 활용해 개인의 취향까지 반영한 유사도를 기반으로 추천을 할 수 있다는 것이죠.


사실 강의를 진행하면서 해당 코드를 돌려봤는데.... 제 컴퓨터로는 30분이 지나도 안 되더군요 ㅠㅠ 
뭐가 이렇게 많이 걸리는지....
때문에 실습은 어려울 것같고, 위의 수식을 함수로 옮겨 적은 것과 어떻게 반환하는지
코드만 적어두도록 하겠습니다.
````

````python
def predict_rating(ratings_arr, item_sim_arr) :
    ratings_pred = ratings_arr.dot(item_sim_arr)
    np.array([np.abs(item_sim_arr).sum(axis=1)])
    return ratings_pred
````

````python
잠재 요인 협업 필터링 추천 시스템

(Latent Factor Collaborative Filtering Recommend System)
최근 들어 가장 많이 쓰이고 있는 방법이 바로 잠재 요인 협업 필터링 추천 시스템입니다.
넷플릭스가 이 방식을 사용하게 되면서 다른 기업들도 해당 방법을 사용중이죠. 



해당 기법의 가장 핵심이 되는 잠재요인(Latent Factor)이라는 단어가 굉장히 익숙하지 않나요?
차원축소 파트를 한번쯤 읽어보신 분들은 아~ 그거~ 하시면서 SVD나 NMF라는
단어를 기억하실 수 있으실 겁니다. 

혹시 기억 안 나시는 분들은 아래의 포스팅에 다녀오셔도 될 거같습니다 ㅎㅎ 
요약하자면, 우리가 인수분해를 하듯이, 하나의 큰 행렬을 여러 개의 작은 행렬들로 표현할 수 있는 방법이었습니다.

작은 행렬들로 표현하는 과정에서, 기존에 없던 새로운 column이나 row적 요소가 나오는데,
그걸 바로 Latent Factor, 잠재요인이라고 했죠. 



이 잠재 요인을 적극적으로 활용한 것이 바로 이번에 알아볼 Latent Factor 기반 추천시스템입니다.


저희가 이제 할 거를 그림으로 미리 설명을 드리면 다음과 같습니다.
이런식으로 왼쪽에 있는 Sparse Matrix인 원본 행렬을 가지고 오른쪽의 2개 이상의 Dense Matrix로 분해하는 겁니다. 이런 과정에서 Factor라는 새로운 요소가 나왔는데, 이 새롭게 나온 요소를 Latent Factor라고 하고, 

이 Latent Factor를 가지고 원본 행렬을 재 구성합니다.


재구성하는 방법은 두 밀집 행렬을 말 그대로 내적시켜주면 되고, 
이렇게 내적한 결과로, 원본 행렬에서 비어있는 부분을 예측하는 것이죠. 
굉장히 직관적이면서 굉장히 강력하죠? 



자, 근데 여기서 문제가 발생합니다. Sparse Matrix를 분해시킨다고 했는데,
NaN값이 많은 기존의 원본 행렬을 분해할 수가 있을까요?

네, 없습니다. 때문에, 우리는 Boosting 알고리즘 시간에 배운 정말 강력한 기법인
SGD를 이용하여 빈 칸을 채워준 다음에 SVD를 진행하도록 하겠습니다. 


혹시 SGD가 기억 안 나시는 분은 아래의 또 제 포스팅을 다녀오시면 좋을 거 같습니당 ㅎㅎ
경사하강법이라는 제목으로 소개시켜드렸습니다.
방법은 아래와 같습니다.



1. 원본의 Sparse Matrix를 Dense로 만들기 위해 임의의 두 행렬 P와 Q로 분해시킴.

2. 임의의 값이 채워진 P와 Q를 내적시키면 원본 행렬과 같은 크기의 예측 행렬이 나오는데,

    이 예측 행렬과 원본 행렬의 차이(MSE)가 낮게 나오는 방향으로 P와 Q의 크기를 업데이트 시킴.

3. 2번의 방법을 바로 SGD라고 함. 

4. MSE를 구하는 방법은, 원본 행렬에서 값이 있는 index와 예측 행렬의 해당 index의 Error을 비교함.


한번 코드로 어떻게 구현하는지 살펴보겠습니다. 
````

````python
import numpy as np
## 원본행렬 (Sparse) 생성
R = np.array([[4, np.NaN, np.NaN, 2, np.NaN ],
              [np.NaN, 5, np.NaN, 3, 1 ],
              [np.NaN, np.NaN, 3, 4, 4 ],
              [5, 2, 1, 2, np.NaN ]])

num_users, num_items = R.shape

## 잠재 요인의 수는 3으로 설정. K는 Hyper Parameter이므로 나중에 Grid로 찾기도 함.
K=3

## 두 개의 행렬로 분해한다고 가정. 정규분포를 가진 Random하게 값을 입력해줌 !! 
np.random.seed(1)
P = np.random.normal(scale=1./K, size=(num_users, K))
Q = np.random.normal(scale=1./K, size=(num_items, K))
print("P:",P)
print("Q:",Q)
````

````python
import pandas as pd
from surprise import Reader, Dataset, SVD, accuracy
from surprise.model_selection import train_test_split

ratings = pd.read_csv('./ratings.csv')
## Reader를 통해 데이터를 읽어올 방식을 정의함
reader = Reader(rating_scale = (0.5,5.0)) ## 평점이 0.5~5.0 사이의 값이라는 얘기!!

## DataFrame을 그대로 바로 읽어올 수는 없고, Dataset의 load_from_df 함수를 이용해서 불러와야함
## 다만 요구되는 순서에 맞게 df를 바꿔준 다음에 불러옴
data = Dataset.load_from_df(ratings[['userId','movieId','rating']], reader)

## surprise는 추가로 sklearn과 비슷하게 train_test_split 기능을 제공함 똑같음!! 
trainset, testset = train_test_split(data, test_size=0.25)

## 어떤 방식으로 추천시스템을 구현할 지 정할 수 있음.
## 바로 위에서 배웠던 SVD기반 행렬분해 잠재요인으로 한번 해보겠음 ! 
model = SVD(n_epochs=50, n_factors=50) ## 잠재요인을 50개로 설정
model.fit(trainset)

## 예측함에 있어서는 2가지로 나뉨.

# 전체 데이터 세트에 대해서 추천을 예측할 때 : test
test_preds = model.test(testset)

# 개별 유저에 대해 평점을 예측할 때 : predict
## 단, user와 item 은 str로 입력되어야함. 
## 물론 아직 user가 해당 item에 대해 평가를 내리지 않아야겟쥬?
predict_preds = model.predict(str(196), str(302))

## 결과보기
print('전체 데이터 예측에 대한 rmse \n',accuracy.rmse(test_preds))
print('+++'*30)
print('196번 User에 대한 예측 평점 : ', predict_preds[3])
````

````python
import pandas as pd
from surprise import Reader, Dataset, SVD, accuracy
from surprise.model_selection import train_test_split

ratings = pd.read_csv('./ratings.csv')
reader = Reader(rating_scale = (0.5,5.0)) 
data = Dataset.load_from_df(ratings[['userId','movieId','rating']], reader)
trainset, testset = train_test_split(data, test_size=0.25)
model = SVD(n_epochs=50, n_factors=50)

# 1.위 코드를 활용하여 사용자 아이디 196번이 예측한 아이템 아이디 302번의 예측평점을 구해주세요.
model.fit(trainset)
predict_preds = model.predict(str(196), str(302))
predict_preds[3]
````

````python
# 2. 위의 코드를 활용하여 전체 데이터 예측에 대한 rmse를 구해주세요.

model.fit(trainset)
test_preds = model.test(testset)
accuracy.rmse(test_preds)
````
