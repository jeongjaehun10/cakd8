```python
def top(df, n=5, columns='tip_pct'):
    return df.sort_values(by=columns, ascending=False)[-n: ]# 뒤에서부터 끝까지  sort하면 뒤로 갈수록 커짐
                                                            # 어센딩하면 폴스 작아짐 
```

```python
top(tips) # n이나 칼럼을 안넣으면 디폴트값 5, pct
```

```python
top(tips, n=10)
```

```python
grouped=tips.groupby(['day','smoker'])
grouped.apply(top) # 그룹별 탑 
```

```python
grouped.apply(top, n=2) #apply함수의 아규먼트는 n값은 안들어감 그러나 마치 어플라이 
                        # 함수의 아규먼트인 것 처럼 쓰면 됨
```

```python
grouped.apply(top, n=2, columns='tip')
```
