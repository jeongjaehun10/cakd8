```python
from tensorflow import keras 
from keras import Sequential
from tensorflow.keras import layers
(x_train, y_train), (x_test, y_test) =mnist.load_data() 

x_train, x_test = x_train/255.0, x_test/255.0 # 유의점은 실수로 만들어주어야함 0~1사이로 만듬

x_train = x_train.reshape((60000, 28*28))
x_test = x_test.reshape((10000, 28*28))

model =Sequential([
    layers.Dense(512, activation='relu'), # 512는 아웃풋이다. 
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer="rmsprop", loss="sparse_categorical_crossentropy", metrics='acc')
model.fit(x_train, y_train, epochs=5, batch_size=128)
```

```python
model.evaluate(x_test, y_test)
```

```python
from keras.optimizers.optimizer_experimental import optimizer
from keras import models 
from keras import layers
from keras.datasets import mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

train_images = train_images.reshape((60000, 28*28))
train_images= train_images.astype('float32')/255

test_images = test_images.reshape((10000, 28*28))
test_images= test_images.astype('float32')/255


network = models.Sequential()
network.add(layers.Dense(512, activation='relu', input_shape=(28*28,))) # 입력층과 은닉층 은닉층에는 활성화함수
                                                                        # 음수는 0 /0보다 크면 양수 relu 비선형
                                                # 선형변환이 100개는 선형변환이 1개와 큰 차이없다.
                                                # 비선형으로 네트워크를 쌓을수록 여러데이터 처리가 가능 복잡한 문제해결
network.add(layers.Dense(10, activation='softmax')) # 출력층 softmax는 확률로 나옴 
network.compile(optimizer="rmsprop", loss="sparse_categorical_crossentropy", metrics='acc')
# 시그모이드는 이진분류 , logistregression도 시그모이드 쓴다. 오즈의비 =승산비, 오즈의 비를 구해서 log취해서 s자함수 
#시그모이드함수이다. 
# 가설공간을 만든다는 네트워클를 만드는 것인데 지금은 네트워크가 작음 
# 공간은 layer가 많을 수록 커진다. layer에서는 units 수를 정한다. layers, units를 정한다. 무조건 공간이 클수록 좋지 않다. 
# yolo 
network.fit(train_images, train_labels, epochs=5, batch_size=128) # 에포크 5번 반복하겠다.
```
