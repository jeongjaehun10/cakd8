```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
type(iris) # 번치 타입 딕셔너리와 비슷하다. 거의 같음 
```

```python
iris.items()
```

```python
iris_data=iris['data'] # 키 접근하기
```

```python
d={'apple':'사과', 'banana':'바나나'}
d['apple']
```

```python
iris_label=iris['target'] # label은 이름 
```

```python
train_test_split(iris_data, iris_label) # 75 대 25로 전체적으로는 4조각으로
```

```python
train_test_split(iris_label ) # 두조각을 내는 함수이다.  # 비율을 7.5대 2.5로
```

```python
train_test_split(iris_label, test_size=0.4 ,random_state=123)
```

```python
X_train, X_test, y_train, y_test,=train_test_split(iris_data, iris_label)
# 총 4 조각으로 됨
```

```python
y_test
```

```python
dt_clf =DecisionTreeClassifier() # 클래스가 객체인 pos기 한대 만듬 
```

```python
dt_clf.fit(X_train, y_train) # 객체가 쓸수 있는 메소드 fit 적합한
                             # fit을 했다는 것은 
    # fitting 적합 학습 train 훈련 다 패턴을 찾고 답을 찾음 
```

```python
X_train.shape # 열은 속성 변수 , 로우는 관측치
           # 데이터수는 112명 이고 속성이 4개 
```

```python
y_train.shape # 112개가 한줄로 됨   

# 1,1,1,1   1  4가지정보를 y의 1가지 정보를 맞춤 X가 크고 y가 작음 
#180, 70, 184, 74  7 
# 170, 48, 149, 39 9  어떠한 패턴을 찾음 X는 홍길동 24세 y는 인간 
# 이러한 찾는 것을 Train이라함 
# 이러한 패턴을 우리가 만든 dt_clf에 저장됨 
```

```python
X_test.shape # 38명 다 물어봐서 얘는 어때? 값은 38개나옴 
```

```python
y_pred=dt_clf.predict(X_test) # 패턴저장된거 맞춰봐 4개의 정보를 순서에 맞게 줘야함 
y_pred
# 첫번쨰 얘는 1, 두번쨰얘는 0이고 그럼 
```

```python
y_test # 사실 이게 정답임 위에랑 맞춰봄 
```

```python
y_pred== y_test # 맞았는지 볼수 있음
```

```python
sum(y_pred== y_test) # True는 1이니 총 36개 맞음
```
