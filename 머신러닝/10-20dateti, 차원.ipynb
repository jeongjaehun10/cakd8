```python
다음 회귀 차수를 높일수록 과적합의 문제가 발생한다. o
편한 분산 트레이드 오프에서 상호작용한다. 
regularize 
regular 
L1 규제를 적용하면 영향력이 크지 않은 회귀 계수 값을 0을 변환한다.

분류의 예측값은 카테고리이고 회귀의 예측값은 연속된 숫자이다. 

회귀에서 학습이 얼마나 잘 되었는가를 평가하는 지표는 무엇인가요?

MSE, RMSE, MAE 

MAE(Mean Absolte Error): 시그마(|yi-y|)/n
MSE(Mean Squared Error): 시그마(|yi-y|)^2/n

회귀 예측 핵심은 주어진 피처와 결정 값 데이터 기반에서 학습을 통해 최적의 회귀계수를 찾아내는 것이다. 

L1 릿지, L2 라쏘,엘라스틱넷 

MSE에 루트를 씌운것 RMSE
실제값과 예측값의 차이를 절대값을 변환해 평균한것 MAE
실젝값과 분산 대비 예측값의 분산 비율을 지표로 하며, 1에 가까울수록 예측 정확도 높다.R^2
실제값과 예측값의 차이를 제곱해 평균한것 MSE


다 같은 말이다. 
Cost Function : 비용함수
Loss Function : 손실함수
Objective Function : 목적함수
Error Function : 오차함수 

OLS 기반읜 회귀 계수는 상관관계가 매우 높아 중요한 피처만 남기고 제거하나 규제를 적용한다. 
2개 이상의 변수.. ab 교호작용이다. 
MAE, MSE, RMSE는 숫자가 낮을수록 잘 예측했다고 표현다. 
logistic regression 는 회귀라는 이름이 붙어 있지만, 사실은 분류에 사용되는 선형모델이다. 
릿지회귀는 alpha 값을 계속 증가시키면 회귀계수 값은 작아진다.
편행에 가까운 것은 직선에 가깝다.  분산에 가까운 것은 차수가 높다. 


1. 다중 공선성은 무엇인가요?

회귀 분석에서 사용된 모형의 일부 예측 변수가 다른 예측 변수와 상관 정도가 높아,  데이터 분석 시 부정적인 영향을 미치는 현상을 말한다.

2. 사이킷런에서는 RMSE를 제공하지 않기에 MAE에 제곱근을 씌워서 계산하는 함수를 직접 만들어야 한다.(O/X)

답 : X

3. 높은 편향/ 낮은 분산에서는 과적합이 되기 쉬우며, 낮은 편향/ 높은 분산에서는 과소적합이 되기 쉽다.(O/X)

답: X

4. 규제선형 모델에서 Min(Rss(W)+alpha·||W||^2 ) alpha 값을 크게하여 회귀 계수의 크기를 제어할 수 있다.(O/X)

답: O 

5. L1 규제를 적용하는 라쏘 회귀는  회귀 계수를  감소시켜 0으로 만든 후 제거해 적절한 피처만 포함시키는 이점이 있으며,

 식은  Min(Rss(W)+alpha·||W||)이다.(O/X) 

답: O

릿지회귀는 L2이다. L2 규제를 추가해 알파의 값이 높으면 회귀계수 예측 영향도가 감소된 모델은 무엇인가?

비용함수에 alpha값으로 패널티를 부여해 회귀계수 값의 크기를 감소시켜 과적합 개선방식을 규제라고 한다.

```

```python
먼저 차원의 저주가 무엇인지 알아보겠습니다. 차원축소의 목적이 결국 이런 현상을 해결하기 위함이니까요. 
차원의 저주란 데이터의 차원이 커질 수록 필요한 데이터의 수가 그만큼 증가하게 되고, 
이에 따라 모델의 성능이 저하되는 현상을 말합니다.
우리의 목적은 이 x,y,z를 포함한 데이터들을 대표적인 또 하나의 지표로 새롭게 나타내는 것을 원합니다. 
그게 평균 점수라면 x+y+z를 각각 1/3씩 곱해서 더해주면 되고, 가중치를 가중해준 점수라면 
각 과목에 대한 가중치를 곱해주면 되겠죠.
피쳐 선택은 말 그대로 필요한 피쳐만 선택하는 것입니다.

예를 들어, 어떤 데이터의 피쳐가

a,b,c,d,e,f,g 로 있을 때, 이 중 필요한 피쳐가 a,b 만 있으면 된다고
판단하면 그냥 a,b만 가져다가 쓰는 것이죠.
피쳐 추출은 기존의 피쳐를 살짝 변형하면서 압축합니다. 
피쳐 선택이 그냥 있는 것 "그대로" 골라오는 것이었다면, 피쳐 추출은 있는 것을 
살짝 변형시키면서 압축을 시킵니다. 
예를 들어, 어떤 데이터의 피쳐가
a,b,c,d,e,f,g로 이루어져 있을 때, 피쳐 추출을 하면 새로운 피쳐 A,B 로 이들을 압축하는 것이죠.


PCA는 데이터의 피쳐들 간의 움직이는 정도인 공분산 행렬을 이용한 차원축소 방법입니다.
"가장 큰 분산을 가지는 축을 기반으로 차원을 축소시키는 기법"이라고 대충~ 설명드리고 넘어가도 됩니다.

공분산은 두 변수가 어떤 관계가 있냐~ 를 나타내주는 척도입니다. 예를 들어, 키가 x, 몸무게가 y일 때
cov(x,y) > 0 이 나오면 두 변수는 양의 관계에 있다고 할 수 있는 것이죠.

선형 변환
벡터는 방향을 지니는 화살표라고 생각하시면 됩니다. 스칼라는 그것들의 크기를 말해주는 숫자구요. 
위에서 선형변환이 이루어질 때, 같이 따라서 바뀌는 벡터도 있을 것이지만,
여전히 그 방향을 유지하는 벡터가 있을 것입니다. 그 벡터를 고유벡터라고 합니다. 
X라는 행렬   (X) transpose시킵니다.
 X XT 곱함
    
이렇게 얻은 공분산 행렬을 잘 보시면 가운데 대각선은 스스로에 대한 공분산이므로 
해당 변수의 분산을 의미하고, 나머지 부분은 정상적으로 공분산을 나타냄을 보실 수 있습니다.
(그래서 결론이 대각선은 var()로, 나머지는 cov()로 나타냈죠!) 

공분산 행렬의 특징이라고 하면, feature size * feature size가 될 것이니, 
정방행렬(정사각형모양)이 될 것이고, 대각선을 기준으로 양 쪽이 같은 값을 가질 것입니다.
(a,b와의 관계나 b,a와의 관계나 같은 값을 가지니까요) 이를 대칭행렬이라고 합니다. 
즉, 공분산 행렬은 정방행렬이자 대칭행렬입니다. 

1. 입력데이터의 공분산 행렬을 구함

2. 공분산 행렬의 고유벡터와 고유값을 구함

3. 고유값이 큰 순서대로 (많이 변형이 이루어진 쪽으로) 상위 고유벡터 N개 추출

4. 그 고유벡터들을 이용하여 입력 데이터 차원 축소

PCA를 진행해줄 때는 스케일링 작업을 꼭꼭 해주셔야 합니다

----------------------------------
LDA또한 PCA와 마찬가지로 차원 축소를 하는 방법 중 하나입니다. 다만, PCA와의 차이라고 한다면 클래스 분류까지 용이하게 해준다는 것이죠. 
PCA는 데이터의 클래스들이 어떻게 분포했는지는 상관하지 않고, 데이터들의 분포만 보고 축을 형성했습니다.
그러나 LDA는 분포 보단 클래스를 나누는 것에 좀 더 초점을 맞췄다고 할 수 있겠습니다.
원리는 생각보다 간단합니다.
클래스간 분산과 클래스 내부 분산만 계산해내면 되거든요.

이런 식으로, 클래스 내부에서의 분산은 작게, 클래스 간 분산은 크게 만들어 내는 축을 찾아서 축소시킵니다. PCA는 입력된 데이터의 공분산행렬을 구해서 고유벡터와 고유값을 구해 차원을 축소시키지만, 
LDA는 클래스 내부 분산과 클래스 간 분산 행렬을 생성한 다음 이들의 고유벡터와 고유값을 구해 축소시킵니다.
데이터의 공분산행렬을 쓸지, 클래스 와 관련된 분산을 쓰는 지에 대한 차이라고 생각하시면 될 것 같습니다. 
또한 PCA는 비지도 학습이지만, LDA는 지도학습입니다. 당연하죠,,? 클래스가 필요하니까요!
때문에 fit을 해줄 때, 정답 데이터도 함께 넣어주어야 합니다.
LDA 또한 일을 수행시키기 전에 꼭 스케일링을 해주셔야 합니다 !!!!!

----------------------------------------------------

PCA와 LDA가 정방행렬에 대한 작업이었다면, 정방행렬이 아닌 일반 행렬엔 어떤 방법을 써야할까요? 
피쳐 추출을 하기 위한 또 하나의 방법은 뭐가 있을까요? 그럴때, 아주 유용한 도구인 SVD를 소개합니다!
SVD는 행렬이 어떻게 분해되는가~를 보고 차원을 축소시키는 방법입니다. 
PCA도 마찬가지 이지만 대상이 정방행렬(PCA는 공분산행렬이었죠)뿐만 아니라
m*n과 같은 행렬도 적용이 가능하다는 것이 특징이죠.
행렬의 곱은 곧 선형변환을 의미하기도 합니다.

시그마는 변하는 크기를 의미하신다고 보시면 됩니다.
그럼 다시 A라는 행렬은

AVVT = U∑VT
VV^T=I 이므로
A = U∑VT

이런 식으로 들어온 행렬을 분해시키는데, 들어온 걸 전부 다 분해 시키는 걸 Full SVD라고 합니다. 

근데 이런 Full SVD는 많이 안 쓴다고 합니다. 보시면 오히려 크기가 뚱뚱해지기만 했지... 
뭔가 "축소"됐다는 느낌은 전혀 없거든요.

그래서 불필요한 요소를 제거한 것이 바로 Compact SVD라고 합니다.

들어온 데이터의 피쳐들이 서로 다중공선성이 높은 경우, 시그마의 특이값이 해당 부분은 0으로 나오게 됩니다. 

예를 들어, 첫번째 피쳐가 두번째 피쳐의 2배로 모두 값이 이루어진 경우, 
다중 공선성이 매우 높은 걸로 나오겠죠. (이게 매우 높다고 표현하는 게 맞는지...
다중공선성이라고 하는 게 맞는지... 헷갈리네요..)  아무튼 이런 경우엔 저기 S로 표시된 부분이 0으로
나온다는 것이죠.

그럼 이렇게 하면 정말 원하는 만큼(r 만큼) 축소를 시킨 것이 나올 것입니다.
그럼 이 축소된 것은 무엇을 의미할까요? 축소시킨 r은 잠재요소로 볼 수 있습니다. 
10000개의 피쳐를 r개로 줄인 것이니, 이 r들은 10000개의 피쳐들이 함축된 어떤 것이라고 할 수 있겠죠.
이렇게 축소시키는 방법을 Truncated SVD라고 합니다. 

--------------------------------------------------
NMF는 이름에서도 자기정체성을 뚜렷하게 명시하듯이, 음수를 포함하지 않는 행렬의 행렬분해를 말합니다.
원본 행렬의 값들이 모두 양수면, 간단하게 행렬이 분해될 수 있다는 것이죠. 이렇게 말입니다! 


왼쪽의 4 x 6 행렬은 오른쪽의 4 x 2 행렬과 2 x 6 행렬로 각각 분해되었습니다. 
정확하게는 아니더라도 어느정도 비슷하게 분해가 가능하다는 얘기죠. 
그럼 여기서 분해된 행렬 각각은 무엇을 의미할까요?
만약 제가 4 x 2, 2 x 6으로 분해하라고 명령했다면, 이 2가 의미하는 것은 뭘까요?
당연하게도 잠재요소의 갯수입니다. 제가 줄이고자 하는 차원의 목적 수이죠. 
             행에 잠재요소    열에 잠재요소
000000        00              000000
000000        00              000000      
000000        00 
000000        00 


그럼 각각의 행렬은 무엇을 의미할까요?
왼쪽의 W라는 이름의 긴 행렬은 원본 row가 있고, 새로운 column (잠재요소)가 들어왔습니다.

이는 "원본 행에 대해 잠재요소의 값이 어떻게 되는가?" 를 알 수 있는 행렬이라고 할 수 있겠습니다.

그럼 오른쪽의 H 행렬은 원본 column이 있고, 새로운 row (잠재요소)가 있습니다.

이는 "새로운 잠재요소가 원본 행렬의 어떤 요소로 이루어져 있는가?"를 알 수 있는 행렬입니다.


그럼 당연히 이런 질문이 들어옵니다. 아니 그럼~ 이건 양수일 때만 쓸 수 있는 건데~ 
제약조건이 많은 NMF보다 PCA나 SVD를 이용해서 차원 축소하는 게 더 편하지 않나염? 

음.. 그렇게 생각할 수 있지만, PCA와 SVD는 모두 음수까지 반영해서 데이터가 모두 양수인 경우엔 오히려
NMF가 더 잘 축소를 시킨다고 합니다. 


```
