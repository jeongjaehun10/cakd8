## 긴 형식에서 넓은 형식으로 피벗하기

```python
# 넓은 포멧은 우리가 보기 편하고 아래로 길쭉한 것은 컴퓨터가 편함 
```

```python
melt : wide -> long
```

```python
df = pd.DataFrame({'key': ['foo', 'bar', 'baz'],
                   'A': [1, 2, 3],
                   'B': [4, 5, 6],
                   'C': [7, 8, 9]})
df 
```

```python
df.melt() # variable키값이고 value에 가치가 들어감 
```

```python
df.melt(value_vars=['A','B'])  # value_vars는 키값을 골라줌 
```

```python
df.melt(id_vars=['key'],value_vars=['A','B']) # id_vars는 키값의 이름 보임
```

```python
df.melt(id_vars=['key'], value_vars=['A','B','C'])
 #value_vars : unpivot할 컬럼 지정, None이면 id_vars를 뺀 나머지 컬럼으로 자동 지정
 #var_name : value_var를 대표하는 이름 지정, None이면 variable을 이름으로 함
 # value_name : value값 이름 지정 , None 이면 value로 지정됨
#id_vars : identity variables로 쓸 컬럼 지정(다수 가능), list or tuple 형식
```
